# Indexación de Estructuras

 - La indexación es un concepto clave en las estructuras de datos en Python y se refiere a cómo accedemos a los elementos individuales de una estructura usando índices (para listas y tuplas) o claves (para diccionarios). A continuación, se detallan ejemplos prácticos de cómo funciona la indexación en diferentes estructuras de datos.

### Listas


```python
lista_4 = [10, 20, 30, 40]
```

```python
# Para acceder a los elementos usando las posiciones

print(lista_4[0])   # Accede al primer elemento
print(lista_4[2])   # Accede al tercer elemento
print(lista_4[-1])  # Accede al último elemento
```

```python
# Para modificar un elemento de la lista

lista_4[1] = 25
print(lista_4)
```

### Tuplas

```python
tupla_5 = ('Juan', 'José', 'Avendaño')
```

```python
# Para acceder a los elementos usando posiciones

print(tupla_5[0])  # Primer elemento
print(tupla_5[-2]) # Segundo elemento desde el final
```

- Nota: La tuplas no se pueden modificar

### Diccionarios

- No se usan números, sino que los elementos están asociados a una clave.

```python
comprador = {
    'Nombre': 'Mario',
    'Edad': 39,
    'Ciudad': 'Medellín'
}
```

```python
# Para acceder a los valores usando claves

print(comprador['Nombre'])
print(comprador.get('Edad'))
```

```python
# Para añadir un nuevo par clave-valor

comprador['Profesion'] = 'Coontador'
print(comprador)
```

# Módulos e Importaciones en Python

### En Python, un módulo es simplemente un archivo que contiene definiciones, declaraciones y código que puedes reutilizar. Los módulos ayudan a organizar tu código en diferentes archivos, mejorando la legibilidad y facilitando su mantenimiento.

 ## ¿Qué es un Módulo?
- ### Un módulo puede contener funciones, variables, clases y otros elementos de código. Python proporciona muchos módulos integrados (como math, random, os, etc.), y también permite crear módulos personalizados para adaptarse a las necesidades específicas de tu proyecto.

## Creación de un Módulo Personalizado
- ### Para crear un módulo personalizado, simplemente guarda tu código en un archivo con extensión .py. Por ejemplo, crea un archivo llamado operaciones.py con las siguientes funciones:

```python
def Suma(r, l):
    """Suma dos números."""
    return r + l

def Resta(r, l):
    """Resta el segundo número del primero."""
    return r - l
```

## Cómo Importar Módulos en Python
- ### Para utilizar las funciones definidas en un módulo, debes importarlo en tu código. Hay varias formas de hacerlo:

   ### a. Importación Completa del Módulo
   ### b. Importación específica de funciones
   ### c. Importación con alias

## Uso de Módulos Integrados de Python

- ### Python incluye muchos módulos integrados que puedes usar de inmediato. A continuación, algunos ejemplos:

- Importación de un Módulo sin Alias

```python
import math

print(math.sqrt(9874))
```

- Importación Específica desde un Módulo

```python
from math import sqrt

Solucion = sqrt(52674)
print(Solucion)
```

- Importación con Alias de Módulos Integrados

```python
import math as M

# Usando la función sqrt con el alias del módulo
Solucion_2 = M.sqrt(5725)
print(Solucion_2) 
```

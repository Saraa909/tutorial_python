## Conversión de tipos de datos

```python
# Para convertir un entero a float

entero_1 = 10
float_1 = float(entero_1)
print(float_1, type(float_1))
```

```python
# Para convertir un string a entero (Ojo: solo en casos donde el str sea un número)

str_1 = "26"
entero_2 = int(str_1)
print(entero_2, type(entero_2))
```

```python
# Para convertir un float a string

float_2 = 15.689
str_2 = str(float_2)
print(str_2, type(str_2))
```

## Estructuras de datos

- Python ofrece varias estructuras de datos incorporadas que permiten almacenar colecciones de datos de diferentes maneras.

### Listas

- Son colecciones ordenadas y mutables de elementos

```python
lista_1 = [2, 5, 7, 9, 12]
print(lista_1)
```

```python
# Para agregar elementos a la lista

lista_1.append(15)
lista_1
```

```python
# Para mostrar las posiciones de los elementos de la lista (Empiezan desde 0)

print(lista_1[0])
print(lista_1[3])
print(lista_1[5])
```

```python
# Para concatenar listas

lista_2 = ["nevera", "lavadora", "microondas"]
lista_1 + lista_2
```

```python
# Para elimminar elementos de la lista

lista_2.remove("nevera")
lista_2
```

```python
# Para insertar elementos en una posición de la lista
lista_2.insert(0, "estufa")
lista_2
```

```python
# Para ordenar una lista

lista_3 = [10.8, 24.0, -1.3, 9]
print(f"Forma original: {lista_3}")
lista_3.sort()
print(f"Forma ordenada: {lista_3}")
```

```python

numero_cc = " 1.025.543.567 "
print(f"Forma original: {numero_cc}")
numero_cc_limpio = numero_cc.replace(".", "").strip()
print(f"Forma limpiada: {numero_cc_limpio}")
```

### Tuplas

- Son colecciones ordenadas pero no mutables

```python
tupla_1 = ("hola", 35, False)
tupla_1
```

```python
# Para mostrar las posiciones de los elementos de una tupla (Empiezan desde 0)

print(tupla_1[0])
print(tupla_1[2])
print(tupla_1[1])
```

```python
tupla_2 = (768, "mundo", True)
tupla_2
```

```python
tupla_3 = (tupla_1 + tupla_2)
tupla_3
```

```python
# Es un método para contar el número de elementos repetidos en la tupla

tupla_4 = (0, 7, 8, 0, 4, 0)
tupla_4.count(0)
```

### Sets

- Son colecciones desordenadas que no permiten que hayan elementos repetidos

```python
nros = (0, 2, 3, 3, 5)
set_1 = set(nros)
print(set_1)
```

```python
nros_2 = [1, 0, -3, 2, 5]
set_2 = set(nros_2)
print(set_2)
```

```python
# Algunas operaciones entre conjuntos

print(set_2.intersection(set_1))
print(set_1.union(set_2))
print(set_1.difference(set_2))
```
